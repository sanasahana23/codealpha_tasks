```python
# Import required libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://quotes.toscrape.com/'

response = requests.get(url)
print(response.status_code)  # 200 means success

# Print first 500 characters of the HTML content
print(response.text[:500])
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
# Let's find all quotes on the page
quotes = soup.find_all('div', class_='quote')
print(f"Found {len(quotes)} quotes on the page.")
```

```python
all_quotes = []

for quote in quotes:
    text = quote.find('span', class_='text').get_text()
    author = quote.find('small', class_='author').get_text()
    tags = [tag.get_text() for tag in quote.find_all('a', class_='tag')]

    all_quotes.append({
        'Quote': text,
        'Author': author,
        'Tags': ', '.join(tags)
    })

# Check first 2 quotes
all_quotes[:2]
```

```python
df = pd.DataFrame(all_quotes)
df.to_csv('quotes.csv', index=False)

print("Saved quotes.csv file!")
df.head()  # Show first 5 rows
```

```python
base_url = 'https://quotes.toscrape.com'
page_url = '/'

all_quotes = []

while page_url:
    print(f"Scraping {base_url + page_url} ...")
    response = requests.get(base_url + page_url)
    soup = BeautifulSoup(response.text, 'html.parser')

    quotes = soup.find_all('div', class_='quote')

    for quote in quotes:
        text = quote.find('span', class_='text').get_text()
        author = quote.find('small', class_='author').get_text()
        tags = [tag.get_text() for tag in quote.find_all('a', class_='tag')]

        all_quotes.append({
            'Quote': text,
            'Author': author,
            'Tags': ', '.join(tags)
        })

    next_button = soup.find('li', class_='next')
    if next_button:
        page_url = next_button.find('a')['href']
    else:
        page_url = None

print(f"Scraped total {len(all_quotes)} quotes!")
```

```python
df = pd.DataFrame(all_quotes)
df.to_csv('all_quotes.csv', index=False)
print("Saved all_quotes.csv file with all quotes!")
df.head()
```

```python
# Assuming your data is already loaded in df

def clean_quotes(text):
    # Remove curly quotes “ ”
    text = text.replace('“', '').replace('”', '')
    # Remove any leftover straight double quotes "
    text = text.replace('"', '')
    return text

# Clean the Quote column
df['Quote'] = df['Quote'].apply(clean_quotes)

# Save CSV without extra quoting (default is fine since quotes removed)
df.to_csv('all_quotes_clean_no_quotes.csv', index=False)

print("Saved cleaned CSV without quotes inside the quote text and no wrapping quotes.")
```
